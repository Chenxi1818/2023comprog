```python
# T1
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import numpy as np
# 各取了5000个样本，分别从三个正态总体
s1 = norm.rvs(-20,10**(1/2),size=5000)
s2 = norm.rvs(-5,5**(1/2),size=5000)
s3 = norm.rvs(3,20**(1/2),size=5000)

# x1,x2,x3
plt.figure(figsize=(8,5)) # 创建画布
plt.hist(s1, bins=10,color='r', label='s1',
        alpha=0.5,)
plt.hist(s2, bins=10,color='y', label='s2',
        alpha=0.5)
plt.hist(s3, bins=10,color='b', label='s3',
        alpha=0.5)
plt.title("Three Histograms of Three Normal Distributions", 
          fontsize=14)
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('$x$', fontsize=12)
plt.show()
```

```python
# T2
mx = []
size = 10000

for i in range(size):
    s1 = norm.rvs(-20,10**(1/2))
    s2 = norm.rvs(-5,5**(1/2))
    s3 = norm.rvs(3,20**(1/2))
    s = np.random.choice([s1,s2,s3],p=[0.5,0.2,0.3])
    mx.append(s)
    
plt.figure(figsize=(8,5))
plt.hist(mx, bins=30,color='r', label='histogram',
        alpha=0.5)
plt.title('Mixture of Normals', fontsize=14)
plt.xlabel(r'$X$', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
'''
compared to the  histogram obtained in the last question, i find that sample drawn from distribution
with greater probability has higher frequency in mixed distribution.
'''
```

```python
# T3
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import numpy as np
import latex

mx = []
size = 10000

for i in range(size):
    s1 = norm.rvs(-20,10**(1/2))
    s2 = norm.rvs(-5,5**(1/2))
    s3 = norm.rvs(3,20**(1/2))
    s = np.random.choice([s1,s2,s3],p=[0.5,0.2,0.3])
    mx.append(s)
    
plt.figure(figsize=(10,4))
plt.hist(mx, bins=30,color='r',
        alpha=0.4,density = True)
sns.kdeplot(mx,label='Kernel density estimate')
sns.rugplot(mx,color='grey')

x = np.linspace(-40,+20,5000)
px1 = norm.pdf(x,scale=10**(1/2),loc=-20)*0.5
px2 = norm.pdf(x,scale=5**(1/2),loc=-5)*0.2
px3 = norm.pdf(x,scale=20**(1/2),loc=3)*0.3
y = px1+px2+px3
plt.plot(x,px1,label='$0.5×\phi(x;-20,10)$')
plt.plot(x,px2,label='$0.2×\phi(x;-5,5)$')
plt.plot(x,px3,label='$0.3×\phi(x;3,20)$')
plt.plot(x,y,label='$p(x)$')

plt.title('Mixture of Normals (N=5000)', fontsize=14)
plt.xlabel(r'$X$', fontsize=12)
plt.legend(loc='best')
plt.show()
```

```python
# T5
import numpy as np
t = sympy.Symbol('t')
def integ(x):
    if x >= -1 and x <= 0:
        return sympy.integrate((15*(1-t**2)**2)/16*(3*(1-(x-t)**2)/4),(t,-1,x+1))
    elif x >= 0 and x <= 1:
        return sympy.integrate((15*(1-t**2)**2/16)*(3*(1-(x-t)**2)/4),(t,x-1,1))

list(map(integ,list(np.linspace(-1,1,500))))
```

# note:
1. 当需要重复使用某个方法时，def函数，欧克？
2. 你这个完全可以def函数，来表示这样一个数学表达式


```python
# T6
import matplotlib.pyplot as plt
import numpy as np
import sympy
t = sympy.Symbol('t')
x = np.linspace(-1,1,500)
def m1(x):
    return 15*(1-x**2)**2/16
m1 = list(map(m1,x))
plt.plot(x,m1,label='$M1(x)$')
def m2(x):
    return 3*(1-x**2)/4
m2 = list(map(m2,x))
plt.plot(x,m2,label='$M2(x)$')
def m12(x):
    if x >= -1 and x <= 0:
        return sympy.integrate((15*(1-t**2)**2/16)*(3*(1-(x-t)**2)/4),(t,-1,x+1))
    elif x >= 0 and x <= 1:
        return sympy.integrate((15*(1-t**2)**2/16)*(3*(1-(x-t)**2)/4),(t,x-1,1))
m12 = list(map(m12,x))
plt.plot(x,m12,label='$M1*M2(x)$')
plt.title('Convolution on Two Functions')
plt.xlabel('x')
plt.ylabel('corresponding value')
plt.legend()
plt.show
```
